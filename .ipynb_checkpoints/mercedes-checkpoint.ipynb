# Introdução

Desde o primeiro automóvel, o Benz Patent Motor Car, em 1886, a Mercedes-Benz tem sido sinônimo de importantes inovações automotivas. Entre elas estão, por exemplo, a célula de segurança para passageiros com zona de deformação, o airbag e os sistemas de assistência inteligentes. A Mercedes-Benz registra cerca de 2.000 patentes por ano, tornando a marca líder europeia entre os fabricantes de automóveis premium. Os carros Mercedes-Benz da Daimler são líderes na indústria de automóveis premium. Com uma enorme seleção de recursos e opções, os clientes podem escolher o Mercedes-Benz personalizado dos seus sonhos.

Para garantir a segurança e a confiabilidade de cada configuração exclusiva de carro antes de eles chegarem às ruas, os engenheiros da Daimler desenvolveram um sistema de testes robusto. No entanto, otimizar a velocidade do sistema de testes para tantas combinações possíveis de recursos é complexo e demorado sem uma abordagem algorítmica poderosa. Como um dos maiores fabricantes mundiais de carros premium, a segurança e a eficiência são fundamentais nas linhas de produção da Daimler.

![carro mercedes](imagens/daimler-mercedes-V02.jpg)

Nesta competição, a Daimler desafia os participantes do Kaggle a enfrentar o problema da dimensionalidade e reduzir o tempo que os carros passam na bancada de testes. Os competidores trabalharão com um conjunto de dados que representa diferentes permutações das características dos carros Mercedes-Benz para prever o tempo necessário para passar nos testes. Os algoritmos vencedores contribuirão para testes mais rápidos, resultando em menores emissões de dióxido de carbono, sem reduzir os padrões da Daimler.

### Validação

As submissões são avaliadas com base no valor R^2(R ao quadrado), também chamado de **coeficiente de determinação**.

### Resultado obtido

Para cada “ID” no conjunto de testes, você deve prever a variável “y”. O arquivo deve conter um cabeçalho e ter o seguinte formato:

```python
ID, y
1,100  
2,100.33  
3,105.81  
...
```

# Introdução ao Dataset

Este conjunto de dados contém um conjunto anônimo de variáveis, cada uma representando uma característica personalizada em um carro Mercedes. Por exemplo, uma variável pode ser tração nas quatro rodas, suspensão a ar adicionada ou um display head-up.

A verdade fundamental é rotulada como “y” e representa o tempo (em segundos) que o carro levou para passar nos testes para cada variável.

### Descrições dos arquivos

As variáveis com letras são categóricas. As variáveis com 0/1 são valores binários.

- ***train.csv:*** O conjunto de treinamento.
- ***test.csv:*** O conjunto de teste, você deve prever a variável “y” para os “IDs” neste arquivo.
- ***sample_submission.csv:*** Um arquivo de envio de amostra no formato correto.
